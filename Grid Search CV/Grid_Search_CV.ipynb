```python
from sklearn.datasets import load_digits
from sklearn.model_selection import GridSearchCV
import numpy as np
import pandas as pd
```

```python
a = load_digits()
df = pd.DataFrame(a.data,columns=[f"pixel_{i}"for i in range(1,65)])
df["target"] = a.target
df
```

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
```

```python
a = [svm.SVC(gamma="auto"),RandomForestClassifier(),LogisticRegression(solver="liblinear"),GaussianNB(),MultinomialNB(),DecisionTreeClassifier()]
params = [{'C': [1,10,20],
'kernel': ['rbf','linear']},
{'n_estimators': [1,5,10]},
{'C': [1,5,10]},{},{},{}]
b = []
for i in range(6):
    cf = GridSearchCV(a[i],params[i], cv = 5 , return_train_score=False)
    cf.fit(df.drop("target",axis = 1),df.target)
    b.append({"model":a[i],"best_score" : cf.best_score_,"param" : cf.best_params_})
k = pd.DataFrame(b)
k
```
