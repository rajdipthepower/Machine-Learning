```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("carprices.csv")
d1 = pd.get_dummies(df["Car Model"])
final=pd.concat([df,d1],axis="columns").drop(["Car Model","Audi A5"],axis="columns")
final
```

```python
plt.scatter(df["Mileage"],df["Sell Price($)"])
plt.ylabel("Sell Price")
plt.xlabel("Mileage")
df.groupby("Age(yrs)")[["Sell Price($)"]].mean().plot(kind="bar")
plt.ylabel("Sell Price")
df.groupby("Car Model")[["Sell Price($)"]].mean().plot(kind="bar")
plt.ylabel("Sell Price")
```

```python
X=final.drop(["Sell Price($)"],axis="columns")
Y=final["Sell Price($)"]
model = LinearRegression()
model.fit(X,Y)
```

```python
model.score(X,Y)
```

```python
#Predict price of a Mercedes Benz that is 4 yr old with mileage 45,000
model.predict([[45000,4,0,1]])[0]
```

```python
#Predict price of a BMW X5 that is 7 yr old with mileage 86,000
model.predict([[86000,7,1,0]])[0]
```
