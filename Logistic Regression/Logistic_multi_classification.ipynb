```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
a = load_iris()
dir(a)
```

a.data contains all the examples (150) of all the flowers having the different values of the features , starting with *setosa , versicolor , virginica*
So , a.data is basically a 2 x 2 array , where first few examples correspond to *Setosa* , next few examples correspond to *versicolor* and last are to *virginica*
And a.target contains 150 values either 0,a or 2 where they correspond to the three flowers in order.

```python
X = a.data
Y = a.target
Xtrain , Xtest , Ytrain , Ytest = train_test_split(X,Y,train_size=0.8)
```

```python
model = LogisticRegression()
model.fit(Xtrain,Ytrain)
```

```python
model.score(Xtest,Ytest)
```

```python
print(model.predict(Xtest))
print(Ytest)
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay

cm = confusion_matrix(Ytest,model.predict(Xtest))
disp = ConfusionMatrixDisplay(confusion_matrix = cm , display_labels = a.target_names)
disp.plot(cmap='Blues')
plt.title("Confusion Matrix")
```
